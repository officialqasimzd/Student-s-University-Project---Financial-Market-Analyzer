## Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px

from ta.volatility import BollingerBands
from ta.momentum import RSIIndicator
from ta.trend import MACD

plt.rcParams['figure.figsize'] = (12,6)
sns.set_style('whitegrid')
```

## Load Dataset

```
df = pd.read_csv('/content/financial_market_dataset.csv')

df['Date'] = pd.to_datetime(df['Date'])

df = df.sort_values('Date')

df.head()
```

## Dataset Information

```
print(df.info())
print(df.describe())
```

## Closing Price Trend Analysis

```
plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Close'])

plt.title('Closing Price Trend')
plt.xlabel('Date')
plt.ylabel('Close Price')

plt.show()
```

## Moving Average Analysis

```
plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['SMA_10'], label='SMA 10')
plt.plot(df['Date'], df['EMA_10'], label='EMA 10')

plt.title('SMA vs EMA Analysis')
plt.legend()

plt.show()
```

## Daily Return Distribution

```
plt.figure(figsize=(10,6))

sns.histplot(df['Daily_Return'], bins=30, kde=True)

plt.title('Daily Return Distribution')

plt.show()
```

## Volatility Analysis

```
plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Volatility'])

plt.title('Market Volatility Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')

plt.show()
```

## Volume Analysis

```
plt.figure(figsize=(14,6))

plt.bar(df['Date'], df['Volume'])

plt.title('Trading Volume Analysis')
plt.xlabel('Date')
plt.ylabel('Volume')

plt.show()
```

## Candlestick Chart

```
fig = go.Figure(data=[go.Candlestick(
    x=df['Date'],
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close']
)])

fig.update_layout(
    title='Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price'
)

fig.show()
```

## Bollinger Bands

```
indicator_bb = BollingerBands(close=df['Close'], window=20, window_dev=2)

df['bb_high'] = indicator_bb.bollinger_hband()
df['bb_low'] = indicator_bb.bollinger_lband()
df['bb_mid'] = indicator_bb.bollinger_mavg()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], df['bb_high'], label='Upper Band')
plt.plot(df['Date'], df['bb_low'], label='Lower Band')
plt.plot(df['Date'], df['bb_mid'], label='Middle Band')

plt.title('Bollinger Bands')
plt.legend()

plt.show()
```

## RSI Indicator

```
rsi = RSIIndicator(close=df['Close'], window=14)

df['RSI'] = rsi.rsi()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['RSI'])

plt.axhline(70, linestyle='--')
plt.axhline(30, linestyle='--')

plt.title('Relative Strength Index (RSI)')

plt.show()
```

## MACD Indicator

```
macd = MACD(close=df['Close'])

df['MACD'] = macd.macd()
df['MACD_Signal'] = macd.macd_signal()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['MACD'], label='MACD')
plt.plot(df['Date'], df['MACD_Signal'], label='Signal')

plt.title('MACD Analysis')
plt.legend()

plt.show()
```

## Correlation Heatmap

```
corr = df[['Open','High','Low','Close','Volume','Daily_Return','Volatility']].corr()

plt.figure(figsize=(10,8))

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Heatmap')

plt.show()
```

## Cumulative Return Analysis

```
df['Cumulative_Return'] = (1 + df['Daily_Return']).cumprod()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Cumulative_Return'])

plt.title('Cumulative Return')

plt.xlabel('Date')
plt.ylabel('Growth')

plt.show()
```

## Rolling Volatility

```
df['Rolling_Volatility'] = df['Daily_Return'].rolling(window=20).std()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Rolling_Volatility'])

plt.title('Rolling Volatility')

plt.show()
```

## Drawdown Analysis

```
rolling_max = df['Close'].cummax()

drawdown = (df['Close'] - rolling_max) / rolling_max

plt.figure(figsize=(14,6))

plt.plot(df['Date'], drawdown)

plt.title('Drawdown Analysis')
plt.ylabel('Drawdown')

plt.show()
```

## Sharpe Ratio

```
sharpe_ratio = (
    df['Daily_Return'].mean() /
    df['Daily_Return'].std()
) * np.sqrt(252)

print('Sharpe Ratio:', sharpe_ratio)
```

## Risk vs Return Scatter

```
plt.figure(figsize=(10,6))

plt.scatter(df['Volatility'], df['Daily_Return'], alpha=0.5)

plt.title('Risk vs Return')
plt.xlabel('Volatility')
plt.ylabel('Daily Return')

plt.show()
```

## Support and Resistance Approximation

```
support = df['Low'].rolling(window=20).min()
resistance = df['High'].rolling(window=20).max()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], support, label='Support')
plt.plot(df['Date'], resistance, label='Resistance')

plt.title('Support and Resistance Levels')
plt.legend()

plt.show()
```
